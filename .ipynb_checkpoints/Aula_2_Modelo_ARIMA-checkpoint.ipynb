## Definindo os Valores de p, d e q

- p: ordem da parte autorregressiva: PACF (função de autocorrelação parcial)
- d: grau de diferenciação: teste de estacionariedade
- q: ordem da média móvel: ACF (função de autocorrelação)

## Escolha dos Modelos

- ***Critério de AIC***: (Menor Melhor) Estima a quantidade relativa de informações perdiadas por determinado modelo.<br> Quanto menos informações um modelo perde, maior a qualidade desse modelo e menor a pontuação AIC.


- ***Critério de BIC***: (Menor Melhor) Critério de Informação Bayesiano

## Simulando Modelo AR(1)

Documentação: https://www.statsmodels.org/dev/generated/statsmodels.tsa.arima_process.ArmaProcess.html

```python
# Criando série aleatória

np.random.seed(1)
pontos = np.random.rand(500)
```

```python
from statsmodels.tsa.arima.model import ARIMA

order = (1,0,0)
ar_coefs = [0.8]

modelo = ARIMA(pontos, 
               order=order, 
               enforce_stationarity=ar_coefs, 
               enforce_invertibility=None).fit()

serie_ar = modelo.fittedvalues
```

```python
# Plotando a série gerada
plt.figure(figsize=(20,8))
plt.plot(serie_ar)
plt.xlabel('Tempo')
plt.ylabel('Valor')
plt.title('Modelo AR(1): X(t) = 0.8.X(t-1) + erro(t)')
plt.show()
```

## Simulando Modelo MA(1)

```python
# Definindo os parâmetros do modelo MA

order = (0,0,1)
ar_coefs = []
ma_coefs = [-0.3]

modelo = ARIMA(pontos, 
               order=order, 
               enforce_stationarity=None, 
               enforce_invertibility=ma_coefs).fit()

serie_ma = modelo.fittedvalues
```

```python
# Plotando a série gerada
plt.figure(figsize=(20,8))
plt.plot(serie_am)
plt.xlabel('Tempo')
plt.ylabel('Valor')
plt.title('Modelo MA(1) X(t)=-0.3erro(t-1) + erro(t)')
plt.show()
```

## Simulando Modelo ARIMA(1,1,1)

Fonte Função: https://lofgrenfinance.wordpress.com/2018/08/25/simulate-an-arima-model-in-python/ <br>
Github: https://github.com/TOMILO87/time_series_simulation/blob/master/ARIMA

```python
#Fonte: https://lofgrenfinance.wordpress.com/2018/08/25/simulate-an-arima-model-in-python/
def func_ARIMA(phi = np.array([0]), theta = np.array([0]), d = 0, t = 0, mu = 0, sigma = 1, n = 20, burn = 10):
    """ Simulate data from ARMA model (eq. 1.2.4):

    z_t = phi_1*z_{t-1} + ... + phi_p*z_{t-p} + a_t + theta_1*a_{t-1} + ... + theta_q*a_{t-q}

    with d unit roots for ARIMA model.

    Arguments:
    phi -- array of shape (p,) or (p, 1) containing phi_1, phi2, ... for AR model
    theta -- array of shape (q) or (q, 1) containing theta_1, theta_2, ... for MA model
    d -- number of unit roots for non-stationary time series
    t -- value deterministic linear trend
    mu -- mean value for normal distribution error term
    sigma -- standard deviation for normal distribution error term
    n -- length time series
    burn -- number of discarded values because series beginns without lagged terms

    Return:
    x -- simulated ARMA process of shape (n, 1)

    Reference:
    Time Series Analysis by Box et al.
    """

    # add "theta_0" = 1 to theta
    theta = np.append(1, theta)
    
    # set max lag length AR model
    p = phi.shape[0]

    # set max lag length MA model
    q = theta.shape[0]

    # simulate n + q error terms
    a = np.random.normal(mu, sigma, (n + max(p, q) + burn, 1))
    
    # create array for returned values
    x = np.zeros((n + max(p, q) + burn, 1))

    # initialize first time series value
    x[0] = a[0]

    for i in range(1, x.shape[0]):
        AR = np.dot(phi[0 : min(i, p)], np.flip(x[i - min(i, p) : i], 0))
        MA = np.dot(theta[0 : min(i + 1, q)], np.flip(a[i - min(i, q - 1) : i + 1], 0))
        x[i] = AR + MA + t

    # add unit roots
    if d != 0:
        ARMA = x[-n: ]
        m = ARMA.shape[0]
        z = np.zeros((m + 1, 1)) # create temp array

        for i in range(d):
            for j in range(m):
                z[j + 1] = ARMA[j] + z[j] 
            ARMA = z[1: ]
        x[-n: ] = z[1: ]
        
    return x[-n: ]
```

```python
ar_coefs = np.array([0.8])
ma_coefs = np.array([-0.3])

serie_arima = func_ARIMA(phi=ar_coefs, theta=ma_coefs, d=1, n=500)
```

```python
# Plotando a série gerada
plt.figure(figsize=(20,8))
plt.plot(serie_arima)
plt.xlabel('Tempo')
plt.ylabel('Valor')
plt.title('Modelo ARIMA(1,1,1)')
plt.show()
```

# Testes de Estacionariedade

### Teste de Dickey-Fuller

Teste mais usado e recomendado para testar se a série é estacionária.

> H0: A série ***não é estacionária***

> H1: A série é estacionária

p-valor < 0,01 rejeita H0

```python
from statsmodels.tsa.stattools import adfuller

# Teste na Série AR
resultado = adfuller(serie_ar)
print(f'stat: {resultado[0]}')
print(f'p_value: {resultado[1]}')
```

```python
# Teste na Série AM
resultado = adfuller(serie_am)
print(f'stat: {resultado[0]}')
print(f'p_value: {resultado[1]}')
```

```python
# Teste na Série ARMA
resultado = adfuller(serie_arma)
print(f'stat: {resultado[0]}')
print(f'p_value: {resultado[1].round(4)}')
```

```python
# Teste na Série ARIMA
resultado = adfuller(serie_arima)

valores_críticos = resultado[4]


print(f'stat: {resultado[0]}')
print(f'p_value: {resultado[1].round(4)}')

print('Valores críticos:')
for chave, valor in valores_críticos.items():
    print('\t', chave, ':', valor)
```

### Teste de KPSS

> H0: A série ***é estacionária***

> H1: A série não é estacionária

p-valor < 0,01 rejeita H0

```python
from statsmodels.tsa.stattools import kpss

valores_críticos = resultado[3]

resultado = kpss(serie_arima)
print(f'stat: {resultado[0]}')
print(f'p_value: {resultado[1]}')

print('Valores críticos:')
for chave, valor in valores_críticos.items():
    print('\t', chave, ':', valor)
```
