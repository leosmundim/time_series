## Definindo os Valores de p, d e q

- p: ordem da parte autorregressiva: PACF (função de autocorrelação parcial)
- d: grau de diferenciação: teste de estacionariedade
- q: ordem da média móvel: ACF (função de autocorrelação)

## Escolha dos Modelos

- ***Critério de AIC***: (Menor Melhor) Estima a quantidade relativa de informações perdiadas por determinado modelo.<br> Quanto menos informações um modelo perde, maior a qualidade desse modelo e menor a pontuação AIC.


- ***Critério de BIC***: (Menor Melhor) Critério de Informação Bayesiano

## Análise da Série Treino

Na Autocorrelação (ACF) não enchergamos o Decaimento Exponencial <br>

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

#Configurações do plot
fig, axes = plt.subplots(3)
plt.gcf().set_size_inches(10, 10)
plt.subplots_adjust(hspace=0.3)

axes[0].plot(train_varejo)
axes[0].set_title('Vendas Base Treino')
plot_acf(train_varejo, ax = axes[1], lags=36);
plot_pacf(train_varejo, ax = axes[2], method='ywm', lags=36);
```

## Teste de Estacionariedade

### Teste de Dickey-Fuller

Teste mais usado e recomendado para testar se a série é estacionária.

> H0: A série ***não é estacionária***

> H1: A série é estacionária

p-valor < 0,01 rejeita H0

```python
from statsmodels.tsa.stattools import adfuller

# Teste na Série TREINO
resultado = adfuller(train_varejo)
print(f'stat: {resultado[0]}')
print(f'p_value: {resultado[1]}')
```

***Coeficiente da primeira defasagem com p-valor de 0,98. Aceita H0, a série não é estacionária***

Ela precisa de uma diferença, será uma série INTEGRADA

### Teste de Estacionariedade para 1 Diferenciação

```python
#Modelo para uma diferenciação já apresenta estacionaridade
res = adfuller(train_varejo.diff().dropna())
print(f'difference first p-value: {res[1]}')
```

#### Teste de KPSS

> H0: A série ***é estacionária***

> H1: A série não é estacionária

p-valor < 0,01 rejeita H0

```python
from statsmodels.tsa.stattools import kpss

#CONFIRMAÇÃO COM TESTE KPSS

resultado = kpss(train_varejo.diff().dropna())
valores_críticos = resultado[3]

print(f'stat: {resultado[0]}')
print(f'p_value: {resultado[1]}')

print('Valores críticos:')
for chave, valor in valores_críticos.items():
    print('\t', chave, ':', valor)
```

## Modelo AUTO-ARIMA

- ***Histogram***: os resíduos estão bem distribuídos para a distribuição normal;
- ***Correlogram***: os resíduos não possuem nenhum lag significativo;
- ***Normal Q-Q***: os valores previstos estão em linha com os valores reais;

```python
arima_varejo.plot_diagnostics();
```

### Teste de Ljung-Box

Referência: https://www.statology.org/ljung-box-test-python/ <br>
Documentação: https://www.statsmodels.org/devel/generated/statsmodels.stats.diagnostic.acorr_ljungbox.html

O teste de Ljung-Box é um teste estatístico que pode ser usado para avaliar a presença de autocorrelação nos resíduos de uma série temporal. <br>
Ele é usado para verificar se as autocorrelações nos resíduos são significativamente diferentes de zero para um determinado número de defasagens.


> **H0: Os resíduos são independentes (modelo não exibe falhas de ajustes)<br>**

> H1: Os resíduos não são independentes (modelo exibe falhas de ajustes)

p-valor < 0.05 rejeita H0 

df no modelo é calculado como p+q (1(p) + 2(q) + 1(P))

```python
from statsmodels.stats.diagnostic import acorr_ljungbox

ljung_box_test = acorr_ljungbox(arima_varejo.resid(), 
                                return_df=True, 
                                lags=[10]) 
ljung_box_test
```

### Teste de Normalidade dos Resíduos

Modelo Aplicado no R apresentou resíduos normais com p-value = 0,04891

### Teste de Verificação da Aderência dos Resíduos à Normalidade (Shapiro-Francia)

Caso os erros não sigam uma distribuição normal, o modelo não é adequado para fins preditivos. <br>
Os betas não são ajustados para que consigam fazer inferência a partir dos intervalos de confiança daqueles parâmetros e construção de cenários.

> Teste Shapiro-Francia: n >= 30

> Teste Shapiro-Wilk: n < 30


onde,

 - H0: Diferênças não são estatisticamente significantes (existe aderência à normalidade);
 - H1: Diferênças são estatisticamente significantes (não é uma normal)
 
p-value < 0,05 rejeita-se H0

```python
from scipy.stats import shapiro

stat, p = shapiro(arima_varejo.resid())
p

#H0: é uma normal
```

```python
# Teste de Normalidade Kolmogorov-Smirnov
from scipy.stats import kstest

kstest(arima_varejo.resid(), 'norm')
```

### Analisando o histograma dos Resíduos e Comparando com o KDE, aparentemente é uma normal

***Para modelos que apresentam resíduos não correlacionados porém que não passam no teste de normalidade, devemos aplicar os modelos heterocedásticos (modelos da família Arch).***

Referência: https://machinelearningmastery.com/develop-arch-and-garch-models-for-time-series-forecasting-in-python/



```python
plt.figure(figsize=(8,8))
sns.histplot(data=arima_varejo.resid(), kde=True, bins=30)
plt.xlabel('Resíduos', fontsize=16)
plt.ylabel('Frequência', fontsize=16)
plt.show()
```

### Teste de Heterocedasticidade de Arch

O teste de Lagrange multiplier, também conhecido como teste LM, é um teste estatístico amplamente usado para avaliar a heterocedasticidade em um modelo de regressão. O p-value do teste de Lagrange multiplier indica a probabilidade de obter um valor estatisticamente significativo do teste, considerando que a hipótese nula seja verdadeira.

Na interpretação do p-value do teste LM, podemos seguir a regra geral:

> H0: Existe heterocedasticidade no modelo;

> H1: Não existe heterocedasticidade no modelo;

p_value < 0,05 rejeita H0

Se o p-value for menor que um nível de significância pré-determinado (geralmente 0,05 ou 0,01), podemos rejeitar a hipótese nula. Isso indica que há evidências estatísticas suficientes para concluir que existe heterocedasticidade no modelo.


Se o p-value for maior que o nível de significância, não temos evidências suficientes para rejeitar a hipótese nula. Nesse caso, não podemos afirmar com confiança que há heterocedasticidade no modelo.

```python
# https://www.statsmodels.org/devel/generated/statsmodels.stats.diagnostic.het_arch.html
from statsmodels.stats.diagnostic import het_arch

teste_arch = het_arch(arima_varejo.resid())
print(f'stat: {teste_arch[0]}')
print(f'p_value: {teste_arch[1]}')
```

***O modelo não é heterocedástico***

## Previsão do Modelo

Documentação: https://alkaline-ml.com/pmdarima/modules/generated/pmdarima.arima.ARIMA.html#pmdarima.arima.ARIMA.predict

```python
prev_varejo = arima_varejo.predict(n_periods=24, 
                                   return_conf_int=True, 
                                   alpha = 0.05) #Intervalo de confiança de 5% (1-alpha)
```

```python
#ax = chpras_fat.chpras_fat_primo.plot(label='observado', figsize=(20,15))

ax = train_varejo.plot(label='Treino')
prev_varejo[0].plot(label= 'Forecast') # previsão do modelo

ax.fill_between(prev_varejo[0].index, # datas
                prev_varejo[1][:,0], # intervalo de confiança inferior
                prev_varejo[1][:,1], # intervalo de confiança superior
                color='blue', alpha=.50)

plt.xlabel('Data')
plt.ylabel('Vendas')
plt.legend()
```

## Acurácia do Modelo - Indicadores de Erros

```python
def func_erros(previsao, teste):

    # calcular as estatísticas de erro
    mae = np.mean(np.abs(previsao - teste))
    mse = np.mean((previsao - teste)**2)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((teste - previsao) / teste)) * 100

    return print('MAE:', mae), print('MSE:', mse), print('RMSE:', rmse), print('MAPE:', mape)
```

```python
func_erros(prev_varejo[0], teste_varejo)
```

# Comparativo com Modelo ETS (Error, Trend, Seasonality)

Documentação: https://www.statsmodels.org/dev/generated/statsmodels.tsa.holtwinters.ExponentialSmoothing.html

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

ets = ExponentialSmoothing(train_varejo, 
                           seasonal_periods=12, #Sazonalidade anual 
                           trend='add', # Tendência Aditiva 
                           damped_trend=True,
                           seasonal='mul').fit() # Sazonal Aditiva
                                            

    
# Faz as previsões para 3 passos à frente
ets_previsao = ets.forecast(steps=24)
ets_previsao = pd.DataFrame(ets_previsao, columns=['y_hat'])
```

#### Modelo ETS apresentou melhores indicadores de acurácia.

```python
func_erros(ets_previsao['y_hat'], teste_varejo)
```

```python
ax = train_varejo.plot(label='Treino')
ets_previsao['y_hat'].plot(label= 'ETS Forecast') # previsão do modelo
teste_varejo.plot(label='Valores Reais')

plt.xlabel('Data')
plt.ylabel('Vendas')
plt.legend()
```
