```python
# Importando a Base ambev

energia = pd.read_excel(r'bases\energia.xlsx')
#energia = energia.set_index('Data')
energia
```

```python
energia['Data'] = pd.date_range(start='1979-01-01', periods=529, freq='M')
energia = energia.set_index('Data')
energia
```

```python
plt.style.use('ggplot')
energia.plot()
```

```python
train_energia = energia['1979-01-31':'2020-12-31']
teste_energia = energia['2021-01-31':'2023-01-31']
```

```python
train_energia
```

```python
teste_energia
```

## Previsão Suavização de Holt-Winters

```python
from statsmodels.tsa.api import ExponentialSmoothing

holt_winters_energia = ExponentialSmoothing(train_energia, 
                                           seasonal_periods=12, #Sazonalidade anual
                                           trend='add', # Tendência Aditiva
                                           seasonal='mul',
                                            damped_trend=True,
                                            freq='M').fit() # Sazonal Aditiva
                                            

    
holt_winters_energia.summary()
```

```python
# Faz as previsões para 25 passos à frente
holt_winters_previsao = holt_winters_energia.forecast(steps=25)
holt_winters_previsao = pd.DataFrame(holt_winters_previsao, columns=['y_hat'])
holt_winters_previsao
```

### MAPE

```python
mape = np.mean(np.abs((teste_energia.Energia - holt_winters_previsao['y_hat']) / teste_energia.Energia)) * 100
mape
```

## Gráfico com os Valores Previstos

```python

# plotar o gráfico da série temporal e dos valores previstos com intervalo de confiança
plt.figure(figsize=(10, 6))
plt.plot(train_energia.index, train_energia, label='Reais')
plt.plot(holt_winters_previsao.index, holt_winters_previsao, label='Valores Previstos')
#plt.fill_between(intervalo_conf.index, intervalo_conf.iloc[:, 0], intervalo_conf.iloc[:, 1],
#                 alpha=0.2, color='gray', label='Intervalo de Confiança (95%)')
plt.legend(loc='best')
plt.title('Suavização Exponencial Holt-Winters Multiplicativo')
plt.xlabel('Data')
plt.ylabel('Valores')
plt.show()
```

## Resíduos

```python
holt_winters_energia.resid.plot()
```

## Função de Autocorrelação

Algumas correlações romperam o intervalo de confiança, sendo assim precisamos aplicar o teste de Ljung-box

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(holt_winters_energia.resid)
```

### Teste de Ljung-Box

O teste de Ljung-Box é um teste estatístico que pode ser usado para avaliar a presença de autocorrelação nos resíduos de uma série temporal. <br>
Ele é usado para verificar se as autocorrelações nos resíduos são significativamente diferentes de zero para um determinado número de defasagens.


> **H0: Os resíduos são idd (modelo não exibe falhas de ajustes)<br>**

> H1: Os resíduos não são idd (modelo exibe falhas de ajustes)

p-valor > 0.05 não rejeita H0 (modelo não possui falhas de ajustes)

```python
from statsmodels.stats.diagnostic import acorr_ljungbox

ljung_box_test = acorr_ljungbox(holt_winters_energia.resid, lags=1)
ljung_box_test
```

<b>P-valor > 0.05, o modelo não exibe falhas de ajuste. <br>
Está bem calibrado, os resíduos não são autocorrelacionados e com baixo erro (2,00%)</b>
